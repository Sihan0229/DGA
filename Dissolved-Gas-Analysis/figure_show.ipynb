```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 
data = {
    '': [
        '', '',
        'IEC', 'IEC',
        'Roger', 'Roger',
        'Duval', 'Duval'
    ],
    '': [
        'RandomForestGini', 'RandomForestEntr',
        'RandomForestGini', 'RandomForestEntr',
        'RandomForestGini', 'RandomForestEntr',
        'RandomForestGini', 'RandomForestEntr'
    ],
    'Accuracy': [0.8875, 0.8875, 0.8938, 0.8875, 0.9, 0.89375, 0.88125, 0.8875],
    'Recall': [0.8875, 0.8875, 0.8938, 0.8875, 0.9, 0.89375, 0.88125, 0.8875],
    'F1 Score': [0.8904, 0.8901, 0.8939, 0.8883, 0.9004, 0.8938, 0.8814, 0.8882],
    'Precision': [0.8968, 0.8957, 0.8988, 0.8931, 0.9049, 0.8968, 0.8897, 0.8946]
}

df = pd.DataFrame(data)

#  seaborn 
df_melted = df.melt(id_vars=['', ''], 
                    value_vars=['Accuracy', 'Recall', 'F1 Score', 'Precision'],
                    var_name='', value_name='')

# 
sns.set(style="whitegrid")

# 
plt.figure(figsize=(14, 8))
chart = sns.barplot(
    x='', y='', hue='',
    data=df_melted, ci=None, palette='Set2',
    hue_order=['Accuracy', 'Recall', 'F1 Score', 'Precision']
)

# 
for container in chart.containers:
    chart.bar_label(container, fmt='%.3f', label_type='edge', fontsize=8, padding=2)

plt.title('', fontsize=16)
plt.xlabel('', fontsize=12)
plt.ylabel('', fontsize=12)
plt.ylim(0.85, 0.91)
plt.legend(title='')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from matplotlib.font_manager import FontProperties

# 
font_cn = FontProperties(fname=r"C:\Windows\Fonts\simsun.ttc", size=18)  # 
font_en = FontProperties(fname=r"C:\Windows\Fonts\times.ttf", size=18)   # Times New Roman

# 
data = {
    '': [
        '', '',
        'IEC', 'IEC',
        'Roger', 'Roger',
        'Duval', 'Duval'
    ],
    '': [
        'RandomForestGini', 'RandomForestEntr',
        'RandomForestGini', 'RandomForestEntr',
        'RandomForestGini', 'RandomForestEntr',
        'RandomForestGini', 'RandomForestEntr'
    ],
    'Accuracy': [0.8875, 0.8875, 0.8938, 0.8875, 0.9, 0.89375, 0.88125, 0.8875],
    'Recall': [0.8875, 0.8875, 0.8938, 0.8875, 0.9, 0.89375, 0.88125, 0.8875],
    'F1 Score': [0.8904, 0.8901, 0.8939, 0.8883, 0.9004, 0.8938, 0.8814, 0.8882],
    'Precision': [0.8968, 0.8957, 0.8988, 0.8931, 0.9049, 0.8968, 0.8897, 0.8946]
}

df = pd.DataFrame(data)

#  seaborn 
df_melted = df.melt(id_vars=['', ''], 
                    value_vars=['Accuracy', 'Recall', 'F1 Score', 'Precision'],
                    var_name='', value_name='')

# 
sns.set(style="whitegrid")

# 
plt.figure(figsize=(16, 4))
chart = sns.barplot(
    x='', y='', hue='',
    data=df_melted, ci=None, palette='Set2',
    hue_order=['Accuracy', 'Recall', 'F1 Score', 'Precision']
)

# 
for container in chart.containers:
    chart.bar_label(container, fmt='%.3f', label_type='edge', fontsize=18, padding=2, fontproperties=font_en)

# 
# plt.title('', fontsize=18, fontproperties=font_cn)
plt.xlabel('', fontsize=18, fontproperties=font_cn)
plt.ylabel('', fontsize=18, fontproperties=font_cn)
plt.ylim(0.88, 0.905)

# 
legend = plt.legend(title='', prop=font_en)
legend.get_title().set_fontproperties(font_cn)

plt.xticks(fontproperties=font_cn)
plt.yticks(fontproperties=font_en)
plt.tight_layout()
plt.show()
```
