```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 构造数据
data = {
    '特征': [
        '关键气体法', '关键气体法',
        'IEC三比值法', 'IEC三比值法',
        'Roger比值法', 'Roger比值法',
        'Duval三角法', 'Duval三角法'
    ],
    '模型': [
        'RandomForestGini', 'RandomForestEntr',
        'RandomForestGini', 'RandomForestEntr',
        'RandomForestGini', 'RandomForestEntr',
        'RandomForestGini', 'RandomForestEntr'
    ],
    'Accuracy': [0.8875, 0.8875, 0.8938, 0.8875, 0.9, 0.89375, 0.88125, 0.8875],
    'Recall': [0.8875, 0.8875, 0.8938, 0.8875, 0.9, 0.89375, 0.88125, 0.8875],
    'F1 Score': [0.8904, 0.8901, 0.8939, 0.8883, 0.9004, 0.8938, 0.8814, 0.8882],
    'Precision': [0.8968, 0.8957, 0.8988, 0.8931, 0.9049, 0.8968, 0.8897, 0.8946]
}

df = pd.DataFrame(data)

# 将长表格格式转换为适合 seaborn 的格式
df_melted = df.melt(id_vars=['特征', '模型'], 
                    value_vars=['Accuracy', 'Recall', 'F1 Score', 'Precision'],
                    var_name='评估指标', value_name='得分')

# 设置可视化风格
sns.set(style="whitegrid")

# 创建图
plt.figure(figsize=(14, 8))
chart = sns.barplot(
    x='特征', y='得分', hue='评估指标',
    data=df_melted, ci=None, palette='Set2',
    hue_order=['Accuracy', 'Recall', 'F1 Score', 'Precision']
)

# 为每个柱添加模型标签
for container in chart.containers:
    chart.bar_label(container, fmt='%.3f', label_type='edge', fontsize=8, padding=2)

plt.title('各方法在不同特征提取方式下的模型评估指标比较', fontsize=16)
plt.xlabel('特征提取方法', fontsize=12)
plt.ylabel('得分', fontsize=12)
plt.ylim(0.85, 0.91)
plt.legend(title='评估指标')
plt.tight_layout()
plt.show()
```
